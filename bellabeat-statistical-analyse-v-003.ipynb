> ### *Question 2: Is the daily, hourly and minutely datas are consistent?*

I want to cross-check the daily, hourly and minute tables for consistency. They should contain the same data have been collected different frequencies. I will compare daily total_steps. For this manipulation I will use the Bigquery SQL again. Because especially minute data is big and I do not want to lose any time.

```python
# Getting asteriks for exact p values
def get_asterisks_for_pval(p_val):
    """Receives the p-value and returns asterisks string."""
    if p_val > 0.05:
        p_text = "'not significant'"
    elif p_val < 1e-4:  
        p_text = '****'
    elif p_val < 1e-3:
        p_text = '***'
    elif p_val < 1e-2:
        p_text = '**'
    else:
        p_text = '*'
    
    return p_text


#normality check
for i in ["total_step_daily","total_step_hour","total_step_minute"]:
    stat, p = shapiro(datatolong[i])
    print("Statistics= ",stat,"p= ",p, get_asterisks_for_pval(p))
    # interpret
    alpha = 0.05
    if p > alpha:
        print(i,'Sample looks Gaussian (fail to reject H0)')
    else:
        print(i,'Sample does not look Gaussian (reject H0)')
```

```python
#therefore I have preferred the Parametric repeated measures one-way ANOVA test
res = pg.rm_anova(dv="mean_total_step", within="data", subject="id", data = data_long, detailed=True)
res
```

> ### *Question 8: Is there a difference of the data existence or quality within time?*

We need to use chi-square hypotesis test for making time-through comparisons. And It has some requirements.
<br>For [Chi-square test assumptions](https://www.statology.org/chi-square-test-assumptions/) and [advance reading](https://www.geeksforgeeks.org/python-pearsons-chi-square-test/).
1. Both variables are categorical.
1. All observations are independent.
1. Cells in the contingency table are mutually exclusive.
1. Expected value of cells should be 5 or greater in at least 80% of cells.




<br>


<br>



```python
#chi square test for the comparison weekdays and weekends. 
#Between April 12 - May 12 2016 we have 264 weekend days 759 weekdays. 
# For advance readings https://www.geeksforgeeks.org/python-pearsons-chi-square-test/
datas = [
        daily_data.groupby("weekend").date.count(),
        daily_data[daily_data.status == "Device is totally active"].groupby("weekend").date.count(),
        daily_data[daily_data.status == "Pedometer and Gps are not active"].groupby("weekend").date.count(),
        ]
names = ['"Data days."','"Totally active days"', '"Days pedometer and gps are inactive"']
for data,name in zip(datas,names):
    data_days = [[data.loc[0], (759 - data.loc[0])],[data.loc[1], (264-data.loc[1])]]
    stat, p, dof, expected = chi2_contingency(data_days)
    alpha = 0.05
    print("H0 is: ", name," is not dependent to part of the week.")
    print("p value is " + str(p))
    get_asterisks_for_pval(p)
    if p <= alpha:
        print("RESULT: ", name,' is dependent (reject H0)')
    else:
        print("RESULT: ", name, ' is independent (H0 holds true)')
```

I have reached Chisquare post-hoc tests [from](https://neuhofmo.github.io/chi-square-and-post-hoc-in-python/) here.

```python
#CHI-SQUARE POST-HOC TESTS

def chisq_and_posthoc_corrected(df):
    """Receives a dataframe and performs chi2 test and then post hoc.
    Prints the p-values and corrected p-values (after FDR correction)"""
    # start by running chi2 test on the matrix
    chi2, p, dof, ex = chi2_contingency(df, correction=True)
    print(f"Chi2 result of the contingency table: {chi2}, p-value: {p}")
    
    # post-hoc
    all_combinations = list(combinations(df.index, 2))  # gathering all combinations for post-hoc chi2
    p_vals = []
    print("Significance results:")
    for comb in all_combinations:
        new_df = df[(df.index == comb[0]) | (df.index == comb[1])]
        chi2, p, dof, ex = chi2_contingency(new_df, correction=True)
        p_vals.append(p)
        # print(f"For {comb}: {p}")  # uncorrected

    # checking significance
    # correction for multiple testing
    reject_list, corrected_p_vals = multipletests(p_vals, method='fdr_bh')[:2]
    for p_val, corr_p_val, reject, comb in zip(p_vals, corrected_p_vals, reject_list, all_combinations):
        print(f"{comb}: p_value: {p_val:5f}; corrected: {corr_p_val:5f} ({get_asterisks_for_pval(p_val)}) reject: {reject}")
```

```python
data = heartrate[(heartrate.id == "15") & ((heartrate.time >= "2016-05-07 12:20:40+00:00") & (heartrate.time <= "2016-05-07 12:21:25+00:00"))]
data = data[["time","value"]]
data = data.set_index("time")["value"].resample("10ms").interpolate(method="linear").reset_index()
data["value"] = data["value"]  / 60
data["t"] = np.arange(len(data))

x_samples = data.t
freq_samples = data.value

x = np.arange(0,4500,0.1)

dx = np.full_like(x, 0.001)       # Change in x

interpolation = interp1d(x_samples, freq_samples, kind='quadratic')
freq = interpolation(x)

x_plot = (freq * dx ).cumsum()    # Cumsum freq * change in x
y = np.sin(x_plot*np.pi*2) + 1
peaks, _ = find_peaks(y, height=0.75)
plt.figure(figsize=(30,8))
plt.plot(x, y, label="pulse")
plt.plot(x, freq, label="freq(x)")
plt.plot(peaks/10, y[peaks], "or", label="peaks")
plt.title("Generated graph of the Device '15' Photoplethysmography (PPG) between on 2016-05-07 12:20:40 and 12:21:25 ")
plt.legend()
#plt.savefig('my_plot.pdf')
plt.show()
```

```python
def extract_appropriate_day():
    list_days = pd.DataFrame()
    for id in heartrate.id.unique():
        ind = heartrate[heartrate.id == id]
        datas = pd.DataFrame()
        for date in ind.time.dt.date.unique():
            day = ind[ind.time.dt.date == date]
            start = day.time.min()
            end = day.time.max()
            if ((end - start) > dt.timedelta(days=0,hours=23,minutes=59)) and (day.time.count() > 8640):
                data = pd.DataFrame({"id":day.id.iloc[0], "date":date, "record_counts":day.time.count(), "time_period":(end-start)}, index = [0])
                datas= pd.concat([data,datas], ignore_index=True)
            else:
                continue
        if len(datas) >= 1:
            list_days = list_days.append(datas.sort_values("record_counts", ascending=False).iloc[0], ignore_index=True)
        else:
            continue
    return list_days
```

```python
def time_basis_analysis(data):
    data = data[["time","value"]]
    data = data.set_index("time")["value"].resample("10ms").interpolate(method="linear").reset_index()
    data["value"] = data["value"]  / 60
    data["t"] = np.arange(len(data))
    x_samples = data.t
    freq_samples = data.value

    x = np.arange(0,8640000,0.1)

    dx = np.full_like(x, 0.001)       # Change in x

    interpolation = interp1d(x_samples, freq_samples, kind='quadratic',fill_value='extrapolate')
    freq = interpolation(x)

    x_plot = (freq * dx ).cumsum()    # Cumsum freq * change in x
    y = np.sin(x_plot*np.pi*2)
    peaks, _ = find_peaks(y, height=0.75) # Peak detection
    nni = tools.nn_intervals(peaks/1000) # RR interval calculation
    results = td.time_domain(nni, plot = False) # Time basis Heartrate variability calculations
    # Access SDNN value using the key 'sdnn'
    res = pd.DataFrame({"hr_mean" : results["hr_mean"], "hr_max":results["hr_max"], "hr_min" : results["hr_min"],
                  "hr_std": results["hr_std"],"sdnn":results["sdnn"],"sdann":results["sdann"],"sdnn_i":results["sdnn_index"],"rmssd":results["rmssd"],
                  "nn50":results["nn50"],"pnn50":results["pnn50"],"tri_index":results["tri_index"] }, index=[0])
    return res
```

* SDNN: Standart deviation of the IBI of normal sinus beats. Below 50ms unhealthy, 50-100ms compromised health, above 100 ms healthy
* SDANN: It is standart deviation of the 5-min segments of the whole 24 h. It does not provide additional information to SDNN.

* NN50: The number of adjacent NN intervals that differ from each other by more than 50 ms (NN50) requires a 2 min epoch.
* pNN50: It is closely correlated parasympathetic nervous system activity, RMSSD and frequency domain HF metric. It is providing an assessment of respiratory sinus arrhythmia with RMSSD.
* RMSSD: The root mean square of successive differences between normal heartbeats (RMSSD) is obtained by first calculating each successive time difference between heartbeats in ms.
* HRV Triangular Index: The HTI is a geometric measure based on 24 h recordings which calculates the integral of the density of the RR interval histogram divided by its height. HTI ≤ 20.42 and RMSSD ≤ 0.068, the heart rhythm is normal. When HTI > 20.42, the pattern is arrhythmic.

I will use the Kaplan-Meier estimate of survival probability formulato analyse device usage and factors effect usage. For further reading about survival analysis. [[1]](https://www.kdnuggets.com/2020/07/complete-guide-survival-analysis-python-part1.html)
$$
S(t) = \prod_{t_i \leq t} (1 - \frac{d_i}{n_i})
$$

- $t_i$ are the events observed in the dataset 
- $d_i$ is the number of deaths at time $t_i$
- $n_i$ is the number of people who we know have survived up to time $t_i$.
